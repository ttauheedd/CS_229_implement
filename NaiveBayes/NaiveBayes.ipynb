```python
import numpy as np
```

```python
import pandas as pd
```

```python
df= pd.read_csv("Naive-Bayes-Classification-Data.csv")
```

```python
x= df.iloc[:,:2]
```

```python
y= df.iloc[:,-1]
```

```python
x['glucose'] = pd.cut(x['glucose'], bins=10, labels=range(1, 11), include_lowest=True)
x['bloodpressure'] = pd.cut(df['bloodpressure'], bins=10, labels=range(1, 11), include_lowest=True)
```

```python
x_train= x.sample(frac=0.8, random_state=42)
y_train= y.sample(frac=0.8, random_state=42)
```

```python
x_test=x.drop(x_train.index)
y_test=y.drop(y_train.index)
```

```python
x_train = x_train.reset_index(drop=True)
x_test  = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test  = y_test.reset_index(drop=True)
```

```python
x_train['glucose'] = x_train['glucose'].astype(int)
x_train['bloodpressure'] = x_train['bloodpressure'].astype(int)
x_test['glucose'] = x_test['glucose'].astype(int)
x_test['bloodpressure'] = x_test['bloodpressure'].astype(int)
```

```python
phi= y_train.mean()
```

```python
glucose_counts =pd.crosstab(x_train['glucose'], y_train)
bloodpressure_counts =pd.crosstab(x_train['bloodpressure'], y_train)
```

```python
class_counts = y_train.value_counts()
```

```python
glucose_theta = (glucose_counts + 1) / (class_counts + 10)
bp_theta = (bloodpressure_counts + 1) / (class_counts + 10)
```

```python
g0 = x_test['glucose'].map(glucose_theta[0])
g1 = x_test['glucose'].map(glucose_theta[1])
```

```python
bp0 = x_test['bloodpressure'].map(bp_theta[0])
bp1 = x_test['bloodpressure'].map(bp_theta[1])
```

```python
prob_0 = (1 - phi) * g0 * bp0
prob_1 = phi * g1 * bp1
```

```python
y_res = (prob_1 > prob_0).astype(int)
```

```python
accuracy = (y_res == y_test).mean()
print(f"Accuracy: {accuracy * 100:.2f}%")
```
