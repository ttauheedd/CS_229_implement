```python
x_train= df.iloc[:,:2]
```

```python
y_train= df.iloc[:,-1]
```

```python
x_train_scaled = (x_train- x_train.mean()) / x_train.std()
```

```python
theta= np.zeros(2)
```

```python
m= len(y_train)
```

```python
phi= sum(y_train)/m
```

```python
k= (y_train==0).sum()
```

```python
l= m-k
```

```python
u0= x_train_scaled[y_train==0].mean(axis=0)
u1= x_train_scaled[y_train==1].mean(axis=0)
```

```python
x_c= x_train_scaled.copy()
```

```python
x_c[(y_train==0)]= x_c[(y_train==0)] - u0
x_c[(y_train==1)]= x_c[(y_train==1)] - u1
```

```python
sigma= (np.dot(x_c.T, x_c))/m
```

```python
sigmainv= np.linalg.pinv(sigma)
```

```python
t0= ((u0 @ sigmainv @ u0) - ( u1 @ sigmainv @ u1))/2 + np.log(phi/(1- phi))
```

```python
t1= sigmainv @ (u1- u0)
```
