```python
x_test=x.drop(x_train.index)
y_test=y.drop(y_train.index)
```

```python
x_train = x_train.reset_index(drop=True)
x_test  = x_test.reset_index(drop=True)
y_train = y_train.reset_index(drop=True)
y_test  = y_test.reset_index(drop=True)
```

```python
x_train_scaled = (x_train - x_train.mean()) / x_train.std()
x_test_scaled = (x_test - x_train.mean()) / x_train.std()
```

```python
m = len(x_train)
X = np.column_stack([np.ones(m), x_train_scaled["x_1"], x_train_scaled["x_2"]])
```

```python
arr_y=np.array(y_train)
```

```python
e= 1e-5
i=0
while np.any(np.abs(theta_prev-theta)>e):
    theta_prev= theta.copy()
    u = np.dot(X, theta)
    h= 1/(1+(np.exp(-u)))
    h = np.clip(h, 1e-15, 1 - 1e-15)
    loss = np.sum(arr_y * np.log(h) + (1 - arr_y) * np.log(1 - h))
    error= h-arr_y
    l_one= np.dot(X.T,error)
    r= h*(1-h)
    l_two= np.dot(X.T*r,X)
    theta = theta- np.linalg.solve(l_two, l_one)
    i=i+1
    
```
